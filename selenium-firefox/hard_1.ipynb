```python
def divide_chunks(l, n):
    # looping till length l
    for i in range(0, len(l), n): 
        yield l[i:i + n]

n = 6 
x = list(divide_chunks(element7, n))
print(x)
df = pd.DataFrame(x,columns =["day date", "Temp High", "Temp Low", "climate", "Humidity", "Wind"])
df
```
