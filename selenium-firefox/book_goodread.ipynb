```python
browser.get('https://www.goodreads.com/list/show/50.The_Best_Epic_Fantasy_fiction_')
print('Title: %s' % browser.title)
```

```python
element1 = browser.find_element_by_class_name("tableList")
```

```python
url_list_a = []
element2 = browser.find_elements_by_tag_name('a')
for elem in element2:
    href = elem.get_attribute('href')
    if href is not None:               # href contains book/
        print(href)
```
